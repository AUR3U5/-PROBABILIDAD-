## ESTADISTICA DESCRIPTIVA DATASET AUTOS

```python
import random
import matplotlib.pyplot as plt
from statistics import*
import pandas as pd
import math
from google.colab import drive
import seaborn as sns
```

```python
autos=pd.read_csv('/content/autos2.csv')
print(autos)
```

##PRECIOS

```python
print("Precio Minimo: ",autos.price.min())
print("Precio Maximo: ",autos.price.max())
m=mode(autos.price)
print("Precio Moda",m)
print("Precios Media: ",autos.price.mean())
l=median(autos.price)
print("Precios Meiana: ",l)
autos = autos.fillna(autos.mean())
d=stdev(autos.price)
print("Desviacion Estandar de Precios: ",d)
print("\n")
autos.describe()

```

```python
plt.hist(autos.price, alpha=0.5, rwidth = 0.95)
plt.title("Precios de Autos")
plt.xlabel("PRECIOS")
plt.ylabel("AUTOS")

n=len(autos.price)
k=int(1+math.log2(n))
stur=pd.cut(autos.price,bins=k).value_counts().reset_index()
stur.columns=("precios","autos")
stur
```

**Conclusiones**:
Podemos observar que la moyoria de los autos
rondan entre los 5000 - 10000, y muy pocos
de ellos alcanzan los $45000

```python
sns.boxplot(x=autos["price"], y=autos["body-style"], data=autos)
plt.title("Arelacion Estadistica entre Precio y Tipo")
plt.ylabel('TIPO')
plt.xlabel('PRECIO')
```

**Conclusiones:** Podemos notar la gran diferencia de precios entre cada tipo
de auto, esto nos puede ayudar a elegir mejor

```python
sns.boxplot(x=autos["price"], y=autos["drive-wheels"], data=autos)
plt.title("Conparativa entre el tipo de rueda del auto y el precio")
plt.ylabel('TRACCION RUEDAS')
plt.xlabel('PRECIO')
```

**Conclusiones:** Podemos notar la gran diferencia de precios, por ejemplo los
que tienen rwd (traccion tracera) requieren mayor potencia como los deportivos, por lo tanto podemos ver que tienen mayor precio, a diferencia de la fwd (traccion delantera) que es la mas comun, por lo tanto mas barata

```python
sns.boxplot(x=autos["price"], y=autos["engine-location"], data=autos)
plt.title("Arelacion Estadistica entre Precio y Ubicacion del Motor")
plt.ylabel('MOTOR')
plt.xlabel('PRECIO')
```

**Cocluciones:** De nuevo podemos notar la gran diferencia de precio, pues los autos con el motor en la parte trasera por lo general suelen ser autos deportivos, por lo tanto, mas caros, como lo vemos en  el diagrama, y los que la tienen en la parte delantera por lo general son mas comunes, mas baratos.
