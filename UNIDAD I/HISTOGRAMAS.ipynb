## HISTOGRAMAS DE 2 GRUPOS
1. En este ejemplo generamos 100 numeros aleatorios reales que representa la produccion diaria de 100 telares (medidas en milimetros)
2. Generar otra lista con la produccion de 100 telares para una segunda fabrica
3. Con esta lista de valores se pretende determinar las medidas de la tendencia central, es decir, la media, la moda, la desviacion estandar.
4. Graficar un histograma para cada fabrica

```
# Libreria que genera numeros aleatorios
import random
# Libreria de estadisticas
from statistics import*
```

```
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
  print(fabrica1)
```

```
fabrica2 = []
for i in range(100):
  numero_aleatorio2 = round(random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio2)
  print(fabrica2)
```

```
#libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```
plt.hist(fabrica1)
plt.title("Produccion Diaria de la Fabrica 1")
```

```
plt.hist(fabrica2)
plt.title("Produccion Diaria de la Fabrica 2")
```

```
media = mean(fabrica1)
mediana = median(fabrica1)
moda = mode(fabrica1)

print("Valores de la fabrica 1")
print("Media:", media)
print("Mediana: ", mediana)
print("Moda: ", moda)
```

```
media = mean(fabrica2)
mediana = median(fabrica2)
moda = mode(fabrica2)

print("Valores de la fabrica 2")
print("Media:", media)
print("Mediana: ", mediana)
print("Moda: ", moda)
```

```
# Generar 2 subplot de 1 renglon por 2 columnas oara desplegar los 2 histogramas,
# una al lado de la otra
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```
#Valor minimo, maximo, rango

media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor="black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Produccion diaria de la fabrica 1")
plt.legend()
plt.show()

```

```
#Valor minimo, maximo, rango

media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

plt.hist(fabrica2, bins=20, alpha=0.2, color='b', edgecolor="black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Produccion diaria de la fabrica 2")
plt.legend()
plt.show()
```
