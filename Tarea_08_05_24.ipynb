```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# par치metros de la distribuci칩n
media = 70  # Media
desv_std = 4  # Desviaci칩n est치ndar
Total_Estudiantes = 500  # Total de estudiantes

Peso_Limite = 72
Peso_Limite_inferior = 68

Prob_menos_68kg = norm.cdf(Peso_Limite, media, desv_std)
Prob_menos_72kg = norm.cdf(Peso_Limite_inferior, media, desv_std)
Prob_entre_68_y_72kg = Prob_menos_68kg - Prob_menos_72kg

Estudiantes_entre_68_y_72kg = Total_Estudiantes * Prob_entre_68_y_72kg

print("Probabilidad de estudiantes que pesan entre 68 y 72 kg:", Prob_entre_68_y_72kg)
print("Cantidad de estudiantes que pesan entre 68 y 72 kg:", Estudiantes_entre_68_y_72kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=((x >= Peso_Limite_inferior) & (x <= Peso_Limite)), facecolor='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese entre 68 y 72 kg. \n  $\mu=70$, $\sigma=4$, x=500   ')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```
